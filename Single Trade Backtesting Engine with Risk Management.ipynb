## Risk Management Overview

Risk management is very crucial in developing robust strategies that are profitable , protect capital and have low drawdowns with good time of recovery.

Risk management methods ->

1) normal take profit 

2) trailing take profit 

3) ATR stop loss 

4) Max drawdown limit to close trades on next day 

5) Intraday Change :Percentage limit of change in close[i]-low[i]/close[i] for long and high[i]-close[i]/close[i] for short trades 

6) Daily Change :Percentage limit of daily change in close price abs(close[i]-close[i-1])/close[i] 

**These risk management methods have greatly help in increasing returns and maintain a low max drawdown**
