```python
def s12(data):
  s12=[]
  sum=0

  for i in range (11):
    s12.append(0)
    sum+=data['Close'].iloc[i]

  sum+=data['Close'].iloc[11]
  s12.append(sum/12)

  for i in range (12,len(data)):
    sum+= data['Close'].iloc[i]
    sum-= data['Close'].iloc[i-12]
    s12.append(sum/12)

  data['s12']=s12

  return data
```

```python
data_prices=s12(data_prices)
data_prices
```

```python
def s26(data):
  s26=[]
  sum=0

  for i in range (25):
    s26.append(0)
    sum+=data['Close'].iloc[i]

  sum+=data['Close'].iloc[25]
  s26.append(sum/26)

  for i in range (26,len(data)):
    sum+= data['Close'].iloc[i]
    sum-= data['Close'].iloc[i-26]
    s26.append(sum/26)

  data['s26']=s26

  return data
```

```python
def e12(data):
  e12=[]

  for i in range (11):
    e12.append(0)

  e12.append(data['s12'].iloc[11])

  f= 2/(12+1)

  for i in range (12,len(data)):
    ema = data['Close'].iloc[i]*f + e12[-1]*(1-f)
    e12.append(ema)

  data['e12']=e12

  return data
```

```python
from math import sqrt

def dev(data):
  dev=[]
  sum=0

  for i in range (24):
    dev.append(0)
    sum+= data['Close'].iloc[i]**2


  sum+=data['Close'].iloc[24]**2
  dev.append(sqrt((sum/25)-data['s25'].iloc[24]**2))

  for i in range (25,len(data)):
    sum+= data['Close'].iloc[i]**2
    sum-= data['Close'].iloc[i-25]**2
    dev.append(sqrt((sum/25)-data['s25'].iloc[i]**2))

  data['dev']=dev

  return data
```
